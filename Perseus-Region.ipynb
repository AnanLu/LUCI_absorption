```python
from Luci import Luci
from Luci.LuciPlotting import plot_spectrum
import matplotlib.pyplot as plt
import numpy as np
```

```python
cube = Luci(cube_dir+'/'+cube_name, cube_dir, 'NGC1275-HR', redshift, 'ML/Reference-Spectrum-R7000', 'ML/R7000-PREDICTOR-I')
#cube = Luci(cube_dir+'/'+cube_name, cube_dir, 'NGC1275-HR', redshift, 'ML/Reference-Spectrum-R7000', model_ML_name = None)
```

```python
axis, bkg_sky = cube.extract_spectrum_region(cube_dir+'/'+'bkg-hires.reg', mean=True)
```
