```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import emcee
from IPython.display import display, Math
import corner
```

```python
# Read in Fits
spectrum = fits.open('Data/Spectrum_1.fits')
header = spectrum[0].header
print(header)
spec = spectrum[1].data
channel = []
counts = []
for chan in spec:  # Only want SN3 region
    channel.append(chan[0])
    counts.append(chan[1])
counts = np.array(counts)#[600:750]
channel = channel#[600:750]
channel_centered = np.linspace(-np.argmax(counts),len(counts)-np.argmax(counts), len(counts))
plt.plot(channel_centered, counts)
plt.show()
```

### Define Model, Log Likelihood, and Prior

```python
def gauss(channel, A, sigma, x0):
    return A*np.exp(-(channel - x0)**2/(2*sigma**2))

def gauss2(channel, A, sigma, x0, A2, sigma2_, x20):
    return A*np.exp(-(channel - x0)**2/(2*sigma**2)) + A2*np.exp(-(channel - x20)**2/(2*sigma2_**2))

def log_likelihood(theta, x, y, yerr, model):
    if model == 1:
        A_, sigma_, x0_ = theta
        model_eval = gauss(x, A_, sigma_, x0_)
    elif model == 2:
        A_, sigma_, x0_, A2_, sigma2_, x20_ = theta
        model_eval = gauss2(x, A_, sigma_, x0_, A2_, sigma2_, x20_)
    sigma2 = 1e-4#yerr ** 2 
    return -0.5 * np.sum((y - model_eval) ** 2 / sigma2 + np.log(2*np.pi*sigma2))

```

```python
A_min = 0
A_max = 4
sigma_min = 0
sigma_max = 5
x0_min = -5
x0_max = 5
A2_min = 0
A2_max = 4
sigma2_min = 0
sigma2_max = 5
x20_min = -5
x20_max = 5
def log_prior(theta, model=1):
    if model == 1:
        A_, sigma_, x0_ = theta
        if x0_min < x0_ < x0_max and sigma_min < sigma_ < sigma_max and A_min < A_ < A_max:
            return 0.0#np.log(1/((x0_max-x0_min)*(sigma_max-sigma_min)*(A_max-A_min)))
    elif model == 2:
        A_, sigma_, x0_, A2_, sigma2_, x20_ = theta
        if x0_min < x0_ < x0_max and sigma_min < sigma_ < sigma_max and A_min < A_ < A_max and x20_min < x20_ < x20_max and sigma2_min < sigma2_ < sigma2_max and A2_min < A2_ < A2_max:
            return 0.0#np.log(1/((x0_max-x0_min)*(sigma_max-sigma_min)*(A_max-A_min)))
    return -np.inf
```

```python
def log_probability(theta, x, y, yerr, model):
    lp = log_prior(theta, model)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, yerr, model)
```

```python
n_walkers = 50
n_dim = 3
init_var = np.array([1.0, 3, 0.1])+ 1e-4 * np.random.randn(n_walkers, n_dim)
init2_var = np.array([1.0, 3, 0.1, 1.0, 3, 0.1])+ 1e-4 * np.random.randn(n_walkers, 2*n_dim)
```

```python
sampler1 = emcee.EnsembleSampler(n_walkers, n_dim, log_probability, args=(channel_centered, counts, 1e-4, 1))
sampler1.run_mcmc(init_var, 2000, progress=True);

sampler2 = emcee.EnsembleSampler(n_walkers, 2*n_dim, log_probability, args=(channel_centered, counts, 1e-4, 2))
sampler2.run_mcmc(init2_var, 2000, progress=True);
```

```python
fig, axes = plt.subplots(n_dim, figsize=(10, 7), sharex=True)
samples = sampler1.get_chain()
labels = ["A", "sigma", "x0"]
for i in range(n_dim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    #ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```
