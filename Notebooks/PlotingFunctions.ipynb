# Plotting Line Functions
In this notebook we plot the line functions for different broadening values


```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.insert(0, '/home/carterrhea/Documents/LUCI/')  # Location of Luci
from LUCI.LuciFunctions import SincGauss, Sinc, Gaussian
```

```python
x_ = np.linspace(0, 1000, 1000)
amp_ = 10
pos_ = 500
broads = [10, 20, 30]
sinc_width = 10  # About 4 orders of magnitude higher than normal value of around 1e-3 or 1e-4
```

```python
fig = plt.figure(figsize=(10,8))
for broad_ in broads:
    plt.plot(x_, np.real(SincGauss(x_, [amp_, pos_, broad_], sinc_width).func), label='%i km/s'%broad_)
plt.legend()
```

```python
fig = plt.figure(figsize=(10,8))
for broad_ in broads:
    plt.plot(x_, np.real(Sinc(x_, [amp_, pos_, broad_], sinc_width).func), label='%i km/s'%broad_)
plt.legend()
plt.xlim(400, 600)
```

```python
fig = plt.figure(figsize=(10,8))
for broad_ in broads:
    plt.plot(x_, np.real(Gaussian(x_, [amp_, pos_, broad_]).func), label='%i km/s'%broad_)
plt.legend()
```
