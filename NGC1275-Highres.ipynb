```python
from LuciBase import Luci
import matplotlib.pyplot as plt
import numpy as np
import LUCI.LuciPlotting as lplt
```

```python
cube_dir = '/media/carterrhea/carterrhea/Benjamin'  # Path to data cube
cube_name = 'NGC1275-highres'  # don't add .hdf5 extension
object_name = 'NGC1275-HR'
redshift = 0.017284
ML_ref = 'ML/Reference-Spectrum-R7000'
ML_model = 'ML/R7000-PREDICTOR-I'
```

```python
cube = Luci(cube_dir+'/'+cube_name, cube_dir, object_name, redshift, ML_ref, ML_model)
```

```python
cube.create_deep_image()
```

```python
vel_map, broad_map, flux_map, chi2_fits = cube.fit_cube(['Halpha', 'NII6548', 'NII6583', 'SII6716', 'SII6731'], 'sincgauss', 950, 1250, 850, 1250, binning=2)
```

```python
lplt.plot_map(vel_map, 'velocity', cube_dir, cube.header, clims=[-500,500])
```
