# Bayesian Analysis

```python
from astropy.io import fits
from scipy import interpolate
import numpy as np
import random
import dynesty
import ipyparallel as ipp
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter("ignore", UserWarning)
from dynesty import plotting as dyplot
import pickle
import time
```

```python
sim_names = ['Single', 'Double']
ref_dir = 'SyntheticData/'#'/media/carterrhea/Dropbox/CFHT/Analysis-Paper2/SyntheticData/'
syn_num = 1000  # Number of Synthetic Data per number of components
```

## Read in Data

```python
# First we have to read in the reference spectra with a resolution of 5000
ref_SN3 = fits.open('Reference-Spectrum-SN3.fits')[1].data
channel = []
counts = []
for chan in ref_SN3:
    channel.append(chan[0])
    counts.append(chan[1])
wavenumbers_syn_SN3 = channel
```

```python
Counts_single = []  # List containing the spectra
Counts_double = []
sim_ct = 0
for ct, sim_name in enumerate(sim_names):
    print('We are on: %s'%sim_name)
    output_dir = ref_dir+sim_name+'/'
    plot_dir = ref_dir+sim_name+'/'
    for spec_ct in range(syn_num):
        if spec_ct%1000 == 0:
            print('  We are on spectrum number %i'%spec_ct)
        spectrum = fits.open(output_dir+'Spectrum_%i.fits'%spec_ct)
        header = spectrum[0].header
        spec = spectrum[1].data
        channel = []
        counts = []
        for chan in spec:
            channel.append(chan[0])
            counts.append(chan[1])
        #channel_centered = np.linspace(-np.argmax(counts),len(counts)-np.argmax(counts), len(counts))#[300:500]
        # interpolate
        f = interpolate.interp1d(channel, counts, kind='slinear')
        # Get fluxes of interest
        coun = f(wavenumbers_syn_SN3[2:-2])  # Need to cut the size just a little bit of the interpolating region otherwise we wont be able to interpolate!
        channel_centered = np.linspace(-np.argmax(coun),len(coun)-np.argmax(coun), len(coun))#[300:500]
        if ct == 0:
            Counts_single.append(np.array(coun))#[300:500])
        elif ct == 1:
            Counts_double.append(np.array(coun))#[300:500])
        else:
            print('Something weird is happening!')
    sim_ct += 1
```

```python
# We need to make the labels
labels = ['single']*syn_num + ['double']*syn_num
# Larger list of combined spectra
spectra = Counts_single+ Counts_double
# Let's now shuffle
# Randomly re-arrange counts/labels
reorder = list(zip(spectra, labels))
random.shuffle(reorder)
spectra, labels = zip(*reorder)
# Quick sanity check
print('Length of spectra and labels: %i %i' % (len(spectra), len(labels)))
```

## Setup Bayesian Testing

```python
# Set training and Test sets
split_num = int(2*syn_num*0.8)  # Number of training to use
Training = spectra[:split_num]
Training_labels = labels[:split_num]
Test = spectra[split_num:]
Test_labels = labels[split_num:]

wic_ratios = []
```

```python
def gauss(x, A, sigma, x0):
    return A*np.exp(-(x - x0)**2/(2*sigma**2))

def gauss2(x, A, sigma, x0, A2, sigma2_, x20):
    return A*np.exp(-(x - x0)**2/(2*sigma**2)) + A2*np.exp(-(x - x20)**2/(2*sigma2_**2))

def loglike1(theta, x, y):
    A_, sigma_, x0_ = theta
    model_eval = gauss(x, A_, sigma_, x0_)
    sigma2 = 1e-4#yerr ** 2 
    return -0.5 * np.sum((y - model_eval) ** 2 / sigma2 + np.log(2*np.pi*sigma2))

def loglike2(theta, x, y):
    A_, sigma_, x0_, A2_, sigma2_, x20_ = theta
    model_eval = gauss2(x, A_, sigma_, x0_, A2_, sigma2_, x20_)
    sigma2 = 1e-4#yerr ** 2 
    return -0.5 * np.sum((y - model_eval) ** 2 / sigma2 + np.log(2*np.pi*sigma2))

# prior transform
def prior_transform1(utheta):
    uA, usigma, ux0 = utheta
    A = 3*uA
    sigma = 10*usigma
    x0 = 200*(ux0)-100

    return A, sigma, x0

# prior transform
def prior_transform2(utheta):
    uA, usigma, ux0, uA2, usigma2, ux20 = utheta
    A = 3*uA
    sigma = 10*usigma
    x0 = 200*(ux0)-100
    A2 = 3*uA2
    sigma2_ = 10*usigma2
    x20 = 200*(ux20)-100

    return A, sigma, x0, A2, sigma2_, x20
```

```python


def bayesian_calc(spectrum, centered_channel):
    # define data to pass to dynesty
    y = spectrum
    x = centered_channel  # np.linspace(-np.argmax(counts),len(counts)-np.argmax(counts), len(counts))
    
    # Calculate logz for single gaussian
    dsampler = dynesty.NestedSampler(loglike1, prior_transform1, ndim=3, logl_args=([x],[y]), sample='rwalk', bound='balls')
    dsampler.run_nested()
    dres = dsampler.results
    ev1 = dres.logz[-1]
    #labels = [r'$A$', r'$sigma$', r'$x0$']
    #fig, axes = dyplot.traceplot(dsampler.results, labels=labels,
                             #fig=plt.subplots(3, 2, figsize=(16, 12)))
    #dsampler.reset()
    # Calcuate logz for double gaussian
    dsampler2 = dynesty.NestedSampler(loglike2, prior_transform2, ndim=6, logl_args=([x],[y]), sample='rwalk', bound='balls')
    dsampler2.run_nested()
    dres2 = dsampler2.results
    ev2 = dres2.logz[-1]
    #dsampler2.reset()

    wic_ratio = ev1/ev2
    wic_ratios.append(wic_ratio)
```

```python
failed = []
passed_ct = 0
failed_ct = 0
start = time.time()
for ct,spectrum in enumerate(Test):
    try:
        bayesian_calc(spectrum, channel_centered)
        passed_ct += 1
    except:
        failed.append(ct)
        failed_ct += 1
end = time.time()
print('Number of passes completed: %i'%passed_ct)
print('Number of passes failed: %i'%failed_ct)
print('Total time: %.2f seconds'%((end-start).total_seconds()))
```

```python
print("To fit %i spectra took %.2f hours"%(passed_ct+failed_ct,(end-start)/3600))
```

```python
# Convert WIC calculations to Model
predictions = []
for ratio in wic_ratios:
    if ratio <= 1.1:
        predictions.append('single')
    else:
        predictions.append('double')
```

```python
print(len(predictions))
```

```python
matrix_conf = confusion_matrix(np.delete(Test_labels,failed),predictions)
# Normalize confusion matrix
matrix_conf = matrix_conf.astype('float64') 
norm_fac = np.sum(matrix_conf[:], axis=1)
for row in range(matrix_conf.shape[0]):
    matrix_conf[row,:] = 100*np.round(matrix_conf[row,:]*(1/norm_fac[row]),3)
# Plot confusion matrix    
sns_plot = sns.heatmap(matrix_conf, annot=True, cmap='Blues')
sns_plot.set(xticklabels=['Single', 'Double'])
sns_plot.set(yticklabels=['Single', 'Double'])
plt.ylabel('True', fontsize='x-large', fontweight='bold')
plt.xlabel('Predicted', fontsize='x-large', fontweight='bold')
#sns.set(font_scale = 1.5)
sns_fig = sns_plot.get_figure()
plt.tight_layout()
plt.savefig('Bayesian.png')
```
